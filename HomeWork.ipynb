```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

N = 210
df_id = np.arange(10001, 10001 + N)
df_score = np.round(np.random.uniform(1.0, 5.0, N), 1)
nan_indices_score = np.random.choice(N, size=int(N * 0.1), replace=False)
df_score[nan_indices_score] = np.nan
df_is_active = np.random.choice([True, False], size=N, p=[0.7, 0.3])
start_date = datetime.now() - timedelta(days=365)
date_range = [start_date + timedelta(days=np.random.randint(365)) for _ in range(N)]
categories = ['Electronics', 'Books', 'Apparel', 'Home Goods', 'Services']
df_category = np.random.choice(categories, size=N, p=[0.25, 0.20, 0.20, 0.15, 0.20])
df_review_count = np.random.randint(0, 500, N)
nan_indices_review = np.random.choice(N, size=int(N * 0.05), replace=False)
df_review_count = df_review_count.astype(float)
df_review_count[nan_indices_review] = np.nan

# --- สร้าง DataFrame ---
data = {
    'Transaction_ID': df_id,                  # Integer
    'Product_Score': df_score,                # Float (มี NaN)
    'Is_Active': df_is_active,                # Boolean
    'Transaction_Date': date_range,           # Date
    'Product_Category': df_category,          # Text/Object
    'Review_Count': df_review_count           # Float ก่อน (มี NaN)
}

df = pd.DataFrame(data)
df['Review_Count'] = df['Review_Count'].astype('Int64')

print(f"DataFrame created: {len(df)} rows, {len(df.columns)} columns.")
df
```
